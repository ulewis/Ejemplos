```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generar una seal senoidal con cambio abrupto de frecuencia
t1 = np.linspace(0, 1, 1000, endpoint=False)
t2 = np.linspace(1, 2, 1000, endpoint=False)
x1 = np.sin(2 * np.pi * 5 * t1)  # Frecuencia de 5 Hz
x2 = np.sin(2 * np.pi * 20 * t2)  # Frecuencia de 20 Hz
x = np.concatenate((x1, x2))

# Aplicar la DWT
niveles = 5
coeficientes = pywt.wavedec(x, 'db4', level=niveles)

# Graficar la seal original y los coeficientes de detalle
plt.figure(figsize=(10, 10))

# Graficar la seal original
plt.subplot(niveles + 1, 1, 1)
plt.plot(x)
plt.title('Seal Original')

# Graficar los coeficientes de detalle
for i, detalle in enumerate(coeficientes[1:]):  # Empezar desde el segundo nivel
    plt.subplot(niveles + 1, 1, i + 2)
    plt.plot(detalle)
    plt.title(f'Detalle Nivel {i+1}')

plt.tight_layout()
plt.show()
```
